```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random

AGENTS = {
    "Windows10-Edge": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0",
    "Windows10-Chrome": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "ChromeOS": "Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36",
    "MacOS": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9",
} # Lista para alternar el user-agent en bucles

URL_BASE = "https://www.lamoncloa.gob.es/consejodeministros/paginas/desclasificacion-documentos-23F.aspx"
```

```python
headers = {"User-Agent": random.choice(list(AGENTS.values()))}
print(f"Procesando {URL_BASE}...")
time.sleep(random.uniform(2, 5))  # Random delay to avoid detection
dd23f = requests.get(URL_BASE, headers=headers, timeout=50, allow_redirects=True)
print(dd23f.status_code)
dd23f_html = dd23f.text
today = pd.Timestamp.today().strftime('%Y_%m_%d')
with open(f'../html/dd23f_html_source_{today}.html', 'w', encoding='utf-8') as file:
    file.write(dd23f_html)
```

```python



soup = BeautifulSoup(dd23f_html, "html.parser")
table = soup.find("table", class_="marketcap-table")
tbody = table.find("tbody")
```
